# 1 Murder or Manslaughter SFO Convictions

```python
# Import Data
Notifications_imported = pd.read_excel("s3://alpha-omppg/FOI/2023 04 19 SFOs Murder or Manslaughter/FOI_230410006_raw_data.xlsx", sheet_name = "All_Notifications")
Notifications_imported.shape
```

```python
# On Licence cases only

SFO_onLicence_Values = ['IPP', 'Life Licence','Post-Release']

onLicenceNotis = Notifications_imported[Notifications_imported["SUPERVISION_TYPE_DESCRIPTION"].isin(SFO_onLicence_Values)].copy()
onLicenceNotis.shape
```

```python
# Outcomes are SFO or death/suicide

SFO_Conviction_Outcomes = ['Death / Suicide', 'Sentenced (SFO)','Hospital Order']

ConvictedOnLicence = onLicenceNotis[onLicenceNotis["OUTCOME_DESCRIPTION"].isin(SFO_Conviction_Outcomes)].copy()
ConvictedOnLicence.shape
```

```python
# Murder or Manslaughter SFO convictions

Offences_Murd_Man = ['Murder', 'Manslaughter','Manslaughter D/R','Child destruction (section 1 of the Infant Life (Preservation) Act 1929)']

Mur_Man_Convs = ConvictedOnLicence[ConvictedOnLicence["CONVICTION_OFFENCE_DESCRIPTION"].isin(Offences_Murd_Man)].copy()
Mur_Man_Convs.shape
```

```python
# Tabulate - conviction offence by year of notification

# Mur_Man_Convs.groupby([Mur_Man_Convs.STAGE_12_DOCN_RECEIVED_ACTUAL.dt.year,"CONVICTION GROUP"])["CONVICTION GROUP"].count().unstack()

Mur_Man_Convs.pivot_table(index = Mur_Man_Convs.STAGE_12_DOCN_RECEIVED_ACTUAL.dt.year, 
                          columns = "CONVICTION GROUP",
                          values = 'SFO_ID',
                          fill_value = 0,
                          aggfunc ='count')
```
