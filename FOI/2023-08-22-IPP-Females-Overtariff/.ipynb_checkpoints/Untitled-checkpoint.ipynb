```python
import numpy as np
import matplotlib.pyplot as plt

# Data
risk = np.array(['Missing', 'Low', 'Medium', 'High', 'Very High'])
risk_score = np.array([484, 9865, 34934, 15717, 449])

# Plotting the pie chart
plt.figure(figsize=(6, 4))
plt.pie(risk_score, labels=risk, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Risk Distribution - Post-Release Probation Caseload 14 Sep 2023')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
risk = np.array(['Missing', 'Low', 'Medium', 'High', 'Very High'])
risk_score = np.array([484, 9865, 34934, 15717, 449])

# Plotting the pie chart
fig, ax = plt.subplots(figsize=(8,4))
wedges, texts, autotexts = ax.pie(risk_score, labels=risk, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Adjust position of 'Missing' and 'Very High' labels
for i, (risk_label, autotext) in enumerate(zip(risk, autotexts)):
    if risk_label == 'Missing':
        autotext.set_position((autotext.get_position()[0], autotext.get_position()[1] + 0.1))
    elif risk_label == 'Very High':
        autotext.set_position((autotext.get_position()[0], autotext.get_position()[1] - 0.1))

    # Adjust position of 'Missing' and 'Very High' text labels
for i, (risk_label, text) in enumerate(zip(risk, texts)):
    if risk_label == 'Missing':
        text.set_position((text.get_position()[0], text.get_position()[1] + 0.08))
    elif risk_label == 'Very High':
        text.set_position((text.get_position()[0], text.get_position()[1] - 0.08))

plt.title('Risk Distribution')
plt.show()
```
