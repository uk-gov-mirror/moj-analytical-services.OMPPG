## Remove duplicates - don't execute code in this section if `rec2` below is empty

```python
# Remove duplicate recalls from recalls data

rec2 = recalls.copy()
rec2['dups'] = rec2.groupby(['FILE_REFERENCE','LICENCE_REVOKE_DATE'])['LICENCE_REVOKE_DATE'].transform('count')
rec2 = rec2[rec2['dups'] > 1]

rec2.shape
```

```python
# Don't execute this if rec2 is empty

rec2['numb1'] = rec2.apply(lambda x: x.str.lower().str.contains("Not Applicable".lower()).sum(), axis=1)
rec2['numb2'] = rec2.apply(lambda x: x.str.lower().str.contains("Not specified".lower()).sum(), axis=1)
rec2['numb3'] = rec2.apply(lambda x: pd.isna(x).sum(), axis=1)
rec2['numb'] = rec2['numb1'] + rec2['numb2'] + rec2['numb3']

rec2.sort_values(['FILE_REFERENCE','numb'], inplace=True)
rec2 = rec2.drop_duplicates(subset=['FILE_REFERENCE'])  # keeps only the first entries with fewer missing data
rec2 = rec2.drop(columns=(['numb1','numb2','numb3','DUPS','numb']))
# rec2[['FILE_REFERENCE','numb']].head(20)
```

```python
rec1 = recalls.copy()
rec1['dups'] = rec1.groupby(['FILE_REFERENCE','LICENCE_REVOKE_DATE'])['LICENCE_REVOKE_DATE'].transform('count')
rec1 = rec1[rec1['dups'] <= 1]
```

```python
print(len(recalls),len(rec1))
```

```python
# Concatenate - don't do this if rec2 has zero entries

recalls1 = rec1.append(rec2).drop(['dups'],axis=1)
recalls1.info()
```

```python
# CHECK AND CORRECT DUPLICATES BY FAMILY NAME, DOB AND LICENCE REVOKE DATE

check1 = recalls.copy()
check1['dups'] = check1.groupby(['FAMILY_NAME', 'DOB', 'LICENCE_REVOKE_DATE'])['LICENCE_REVOKE_DATE'].transform('count')
check1 = check1[check1['dups'] > 1]
check1.shape
```

```python
# 5 Look for excess matches and non-Matches and make corrections

recalls_pub = duckdb.sql("\
    SELECT a.*, b.DOS, b.NUMBER_OF_RECALL_REASONS, \
        b.RECALL_REASON_DESCRIPTIONS,b.TARIFF_EXPIRY_DATE,b.FAMILY_NAME AS EXTRAS_FAMILY_NAME, \
        b.FIRST_NAMES AS EXTRAS_FIRST_NAME,b.DOB AS EXTRAS_DOB,b.FILE_REFERENCE AS FR,\
        b.PRISON_NUMBER AS PN\
    FROM recalls AS a LEFT JOIN extras AS b \
    ON  \
        (a.PRISON_NUMBER = b.PRISON_NUMBER  OR\
         a.FILE_REFERENCE = b.FILE_REFERENCE OR\
         (a.FAMILY_NAME = b.FAMILY_NAME AND a.DOB = b.DOB)\
        ) AND \
        a.LICENCE_REVOKE_DATE = b.LICENCE_REVOKE_DATE").df()

recalls_pub.shape
```

```python
# Excess matches
retain_vars =['FILE_REFERENCE','PRISON_NUMBER', 'FAMILY_NAME', 'FIRST_NAMES','DOB',
              'LICENCE_REVOKE_DATE', 'EXTRAS_FAMILY_NAME', 'EXTRAS_FIRST_NAME','EXTRAS_DOB',
              'RTC_DATE', 'RECALL_TYPE_DESCRIPTION', 'FR','PN']

check1 = recalls_pub.copy()
check1['dups'] = recalls_pub.groupby(['FILE_REFERENCE','LICENCE_REVOKE_DATE'])['LICENCE_REVOKE_DATE'].transform('count')
check1 = check1[check1['dups'] > 1][retain_vars]

check1.shape
```

```python
# Check Non-Matches and resolve if check1 has entries

first_cols = ['FILE_REFERENCE', 'PRISON_NUMBER', 'FAMILY_NAME', 'LICENCE_REVOKE_DATE', 'RTC_DATE', 'RECALL_TYPE_DESCRIPTION']
last_cols = [col for col in recalls_pub.columns if col not in first_cols]

# chdf = df[first_cols+last_cols]
check1 = recalls_pub.loc[recalls_pub.RECALL_REASON_DESCRIPTIONS.isna()][first_cols+last_cols]
display(check1)
```
