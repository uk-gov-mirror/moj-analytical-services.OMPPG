```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

%config InlineBackend.figure_format = 'svg'
```

```python
# Sample data
recalls = pd.read_excel("Recalls.xlsx")
recalls.info()

recalls
```

```python
recalls[20:]['Quarter']
```

```python
# Define color palette
# colors = ["#00B0F0", "lightsteelblue", "#0070C0", "darkviolet"]

#colors = ["steelblue", "#00B0F0", "#0070C0","darkorange"]
# Create a figure and axis object
fig, ax = plt.subplots(figsize=(10, 5))

# Define bar width and offset
#bar_width = 0.3
#offset = 0.15  # Adjusted for overlap
x = recalls['Quarter']
y = recalls['Recalls']
# Plot the total22
# ax.bar(recalls['Quarter'], Recalls["Recalls.x"], width=bar_width, label="Total",color='lightgrey', alpha=0.9,zorder=1)

ax.plot(x, y)

# Only plot Cat1, Cat2, and Cat3 for years up to 2022

# Remove x and y labels for a cleaner look
ax.set_xlabel("")
ax.set_ylabel("")

plt.xticks(rotation=45,ha='right')
# Set x-ticks with bold font and format y-axis as comma-separated
ax.set_xticks(x)
#ax.tick_params(axis='x', which='both', pad=3,color='white')
#ax.set_xticklabels(cats["Year"],fontweight = 'normal',fontsize=11)
ax.set_yticks(np.arange(0,10001,1000))
ax.set_yticklabels(np.arange(0,10001,1000),fontweight = 'normal',fontsize=11)
ax.set_xticklabels(x,fontweight = 'normal',fontsize=8)
#ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{int(x):,}"))

# Customizing the legend
#ax.legend(bbox_to_anchor=(0.025, 0.97), loc="upper left", frameon=False,ncol=3)

# ax.margins(0.02, 0) 
# Hide all axes lines for a minimalistic look
#for spine in ax.spines.values():
    #spine.set_visible(False)

# Add faint reference lines behind bars for context
for y_value in range(1000, 9001, 1000):
    ax.axhline(y=y_value, color="lightgray", linestyle="--", linewidth=1, alpha=0.3, zorder=1)

#plt.savefig("recalls.svg",bbox_inches='tight',pad_inches=0)    
# Display the plot
#plt.show()
```
